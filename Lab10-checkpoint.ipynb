```python
import pyglet
from pyglet import app, gl, graphics
from pyglet.gl import *
from pyglet.window import Window, key
import numpy as np, math
from sys import exit
import random
```

```python
h = 20 # Половины длины ребра куба
w = 2 * h # Для задания области вывода
width = height = 300 # Размер окна вывода

mt_clr0 = (GLfloat * 4)(*[1, 1, 0, 0.45])
mt_clr1 = (GLfloat * 4)(*[0, 1, 1, 0.75])
light_position = (GLfloat * 4)(*[0, 40, 40, 0])
lght_clr = (GLfloat * 4)(*[1, 1, 1, 1])

rot_x = 20
rot_y = 20
rot_x_2 = 20
rot_y_2 = 30
```

```python
verts = ((h, -h, -h), # Координаты вершин куба
         (h, h, -h),
         (-h, h, -h),
         (-h, -h, -h),
         (h, -h, h),
         (h, h, h),
         (-h, -h, h),
         (-h, h, h))
verts_2 = ((h/2, -h/2, -h/2), # Координаты вершин куба
         (h/2, h/2, -h/2),
         (-h/2, h/2, -h/2),
         (-h/2, -h/2, -h/2),
         (h/2, -h/2, h/2),
         (h/2, h/2, h/2),
         (-h/2, -h/2, h/2),
         (-h/2, h/2, h/2))
faces = ((0, 1, 2, 3), # Индексы вершин граней куба
         (3, 2, 7, 6),
         (6, 7, 5, 4),
         (4, 5, 1, 0),
         (1, 5, 7, 2),
         (4, 0, 3, 6))
nrms = ((h, -h, -h),
         (h, h, -h),
         (-h, h, -h),
         (-h, -h, -h),
         (h, -h, h),
         (h, h, h),
         (-h, -h, h),
         (-h, h, h))
```

```python
def cube_draw(mt_clr, verts):
    glMaterialfv(GL_FRONT, GL_SPECULAR, mt_clr)
    glMaterialfv(GL_FRONT, GL_DIFFUSE, mt_clr)
    glMaterialfv(GL_FRONT, GL_AMBIENT, mt_clr)
    for face in faces:
        v4, n4 = (), ()
        for v in face:
            n4 += nrms[v]
            v4 += verts[v]
        graphics.draw(4, gl.GL_QUADS, ('v3f', v4), ('n3f', n4))
```

```python
window = Window(visible = True, width = width, height = height,
        resizable = True, caption = 'Лабораторная работа №10')
glClearColor(1, 1, 1, 1)
glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)
glDisable(GL_DEPTH_TEST)
glDepthFunc(GL_LESS)
glEnable(GL_LIGHTING)
glEnable(GL_LIGHT0)
glLightfv(GL_LIGHT0, GL_SPECULAR, lght_clr)
glBlendFunc(GL_SRC_ALPHA, GL_ONE_MINUS_SRC_ALPHA)
glEnable(GL_BLEND)
@window.event
def on_draw():
    window.clear()
    glMatrixMode(GL_PROJECTION)
    glLoadIdentity()
    glOrtho(-w, w, -w, w, -w, w)
    glMatrixMode(GL_MODELVIEW)
    glLoadIdentity()
    glLightfv(GL_LIGHT0, GL_POSITION, light_position)
    gl.glRotatef(rot_x, 1, 0, 0)
    gl.glRotatef(rot_y, 0, 1, 0)
    cube_draw(mt_clr0, verts)
    gl.glLoadIdentity()
    gl.glRotatef(rot_x_2, 1, 0, 0)
    gl.glRotatef(rot_y_2, 0, 1, 0)
    cube_draw(mt_clr1, verts_2)
```

```python
@window.event
def on_key_press(symbol, modifiers):
    global rot_x, rot_y, rot_x_2, rot_y_2
    if symbol == key._1:
        if modifiers & key.MOD_CTRL:
            gl.glEnable(GL_BLEND)
            gl.glEnable(gl.GL_CULL_FACE)
        else:
            gl.glDisable(GL_BLEND)
    elif symbol == key._2:
        if modifiers & key.MOD_CTRL:
            gl.glEnable(GL_DEPTH_TEST)
        else:
            gl.glDisable(GL_DEPTH_TEST)
    elif symbol == key._3:
        if modifiers & key.MOD_CTRL:
            rot_y -= 5
        else:
            rot_y += 5
    elif symbol == key._4:
        if modifiers & key.MOD_CTRL:
            rot_x -= 5
        else:
            rot_x += 5
    elif symbol == key._5:
        if modifiers & key.MOD_CTRL:
            rot_y_2 -= 5
        else:
            rot_y_2 += 5
    elif symbol == key._6:
        if modifiers & key.MOD_CTRL:
            rot_x_2 -= 5
        else:
            rot_x_2 += 5
    elif symbol == key._7:
        rot_x = 20
        rot_y = 20
        rot_x_2 = 20
        rot_y_2 = 30

app.run()
```
